```python
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
%matplotlib inline
```

```python
chembl_file_path = '../proj1/chembl.csv'
df = pd.read_csv(chembl_file_path, delimiter=",")
df.head()
```

```python
df2 = df[df.standard_value.notna()]
df2
```

```python
bioactivity_class = []
for i in df2.standard_value:
  if float(i) >= 10000:
    bioactivity_class.append("inactive")
  elif float(i) <= 1000:
    bioactivity_class.append("active")
  else:
    bioactivity_class.append("intermediate")
```

```python
selection = ['canonical_smiles', 'molecule_chembl_id', 'standard_value']
df3 = df2[selection]
df3
```

```python
bioactivity_class = pd.Series(bioactivity_class, name='bioactivity_class')
df4 = pd.concat([df3, bioactivity_class], axis=1)
df4
```

```python
from rdkit import Chem
from rdkit.Chem import Descriptors, Lipinski
```

```python
def lipinski(smiles, verbose=False):

    moldata= []
    for elem in smiles:
        mol=Chem.MolFromSmiles(elem) 
        moldata.append(mol)
       
    baseData= np.arange(1,1)
    i=0  
    for mol in moldata:        
       
        desc_MolWt = Descriptors.MolWt(mol)
        desc_MolLogP = Descriptors.MolLogP(mol)
        desc_NumHDonors = Lipinski.NumHDonors(mol)
        desc_NumHAcceptors = Lipinski.NumHAcceptors(mol)
           
        row = np.array([desc_MolWt,
                        desc_MolLogP,
                        desc_NumHDonors,
                        desc_NumHAcceptors])   
    
        if(i==0):
            baseData=row
        else:
            baseData=np.vstack([baseData, row])
        i=i+1      
    
    columnNames=["MW","LogP","NumHDonors","NumHAcceptors"]   
    descriptors = pd.DataFrame(data=baseData,columns=columnNames)
    
    return descriptors
```

```python
df_lipinski = lipinski(df4.canonical_smiles)
```

```python
df_lipinski
```

```python
df_combined = pd.concat([df4,df_lipinski], axis=1)
```

```python
df_combined
```

```python
import numpy as np

def pIC50(input):
    pIC50 = []

    for i in input['standard_value']:
        molar = i*(10**-9) # Converts nM to M
        pIC50.append(-np.log10(molar))

    input['pIC50'] = pIC50
    x = input.drop('standard_value', 1)
        
    return x
```

```python
df_final = pIC50(df_combined)
df_final.head(5)
```

```python
df_2class = df_final[df_final.bioactivity_class != 'intermediate']
df_2class
```

```python
selection = ['canonical_smiles']
df_new = df_2class[selection]
#saving Smiles notation of molecules in smiles format file
df_new.to_csv('smile2mol.smi',sep='\t', index=False, header=False)
```

```python
# preprocessing the collected descriptor data

from sklearn.preprocessing import scale

df1 = df.dropna(axis=1)
df2 = df1.dropna(axis=0)
df2

# scaling the descriptors
threshold = 1000

df3=df2.drop(df2.mean()[df2.mean() > threshold ].index.values, axis=1)
df4=df3.drop(df3.mean()[df3.mean()==0 ].index.values, axis=1)
df5=df4.drop('name',axis=1)
df5.describe()
```

```python
#scaling the values and storing the data

df4_scaled=scale(df5)
df4_scaled_pd=pd.DataFrame(df4_scaled)
df4_scaled_pd.to_csv('descriptors_scaled.csv',index=False)
```
